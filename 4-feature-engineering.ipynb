Filter the dataset to retain only the features required for building prediction models, one-hot encode categorical columns, and cast the entire data frame to float format to prepare it for machine learning workflows.

```python
# !pip install pandas
import pandas as pd
```

```python
df=pd.read_csv("data/dataset_part_2.csv")
```

```python
#Only keep features that will be used for building prediction models
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
# One-hot encode categorical columns in features, creating binary indicators for each unique category
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])
features_one_hot.head()
```

```python
#cast the entire dataframe to float64 datatype
features_one_hot = features_one_hot.astype('float64')
```

```python
#export dataframe as a csv file
features_one_hot.to_csv('data/dataset_part_3.csv', index=False)
```
